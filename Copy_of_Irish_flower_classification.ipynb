```python
import matplotlib.pyplot as plt
```

```python
# Sample data
labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
sizes = [50,50,50]

# Define custom colors
colors = ['orange', 'green', '#32CD32']

# Create a pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors)

# Add a title
ax.set_title('Species Distribution')

# Show the chart
plt.show()
```

```python
from pandas.plotting import scatter_matrix
import seaborn as sns
```

```python
# scatter plot matrix
# Define the colors for each target class
colors = {'Iris-setosa':'red', 'Iris-versicolor':'blue', 'Iris-virginica':'green'}
# Create a scatter matrix plot with colors
scatter_matrix(iris_data, c=iris_data['species'].apply(lambda x: colors[x]))
# Set the title and show the plot
plt.suptitle('Iris Scatter Matrix', size=15)
plt.show()
```
